```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def plot_sine(frequency=1.0):
    x = np.linspace(0, 2 * np.pi, 400)
    y = np.sin(frequency * x)
    
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f"Frequency = {frequency} Hz")
    plt.legend()
    plt.xlim(0, 2 * np.pi)
    plt.ylim(-1.1, 1.1)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Interactive Sine Wave Plot')
    plt.grid(True)
    plt.show()

widgets.interactive(plot_sine, frequency=widgets.FloatSlider(value=1.0, min=0, max=10.0, step=0.1))
```

```python

import numpy as np
import matplotlib.pyplot as plt

class InteractivePlot:
    def __init__(self):
        self.fig, self.ax = plt.subplots(figsize=(8, 6))
        self.ax.set_xlim(0, 10)
        self.ax.set_ylim(0, 10)
        self.ax.set_title("Left click to add point, Right click to remove nearest point")
        self.points, = self.ax.plot([], [], 'bo', markersize=10) # Initial empty plot
        self.x_data = []
        self.y_data = []
        self.cid = self.fig.canvas.mpl_connect('button_press_event', self.onclick)

    def onclick(self, event):
        # Left click adds point
        if event.button == 1:
            self.x_data.append(event.xdata)
            self.y_data.append(event.ydata)
        # Right click removes the nearest point
        elif event.button == 3:
            if len(self.x_data) == 0: return
            nearest_index = np.argmin(np.sqrt((np.array(self.x_data) - event.xdata)**2 + 
                                              (np.array(self.y_data) - event.ydata)**2))
            self.x_data.pop(nearest_index)
            self.y_data.pop(nearest_index)
        self.update_plot()

    def update_plot(self):
        self.points.set_data(self.x_data, self.y_data)
        self.fig.canvas.draw()

plot = InteractivePlot()
```

```python
class BlockImageEditor:
    def __init__(self, blocks_shape, block_size=12):
        self.blocks_shape = blocks_shape
        self.block_size = block_size
        
        # Initialize color picker
        self.color_picker = widgets.ColorPicker(
            value='#0000FF',
            description='Pick a color:'
        )
        self.color_picker.observe(self.on_color_change, names='value')
        
        # Display the color picker
        display(self.color_picker)
        
        # Initialize the image
        self.image = np.zeros((blocks_shape[0]*block_size, blocks_shape[1]*block_size, 3), dtype=np.uint8)
        
        # Plot the image
        self.fig, self.ax = plt.subplots(figsize=(6, 6))
        self.ax.imshow(self.image)
        self.ax.axis('off')
        
        # Connect the onclick event
        self.cid = self.fig.canvas.mpl_connect('button_press_event', self.onclick)
        
        # Set the initial RGB value based on the color picker
        self.rgb = self.hex_to_rgb(self.color_picker.value)
        
    def onclick(self, event):
        # Find the block index
        row, col = int(event.ydata // self.block_size), int(event.xdata // self.block_size)
        self.update_block_color(row, col, self.rgb)
        
    def update_block_color(self, row, col, color):
        start_row, end_row = row*self.block_size, (row+1)*self.block_size
        start_col, end_col = col*self.block_size, (col+1)*self.block_size
        self.image[start_row:end_row, start_col:end_col] = color
        self.ax.imshow(self.image)
        self.fig.canvas.draw()
        
    def hex_to_rgb(self, value):
        value = value.lstrip('#')
        return tuple(int(value[i:i+2], 16) for i in (0, 2, 4))
    
    def on_color_change(self, change):
        self.rgb = self.hex_to_rgb(change['new'])

# Initialize the editor with an image of 5x5 blocks
editor = BlockImageEditor(blocks_shape=(5, 5))
```
